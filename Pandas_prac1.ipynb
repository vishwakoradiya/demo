```python
import numpy as np
import pandas as pd

# Create a sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Ella'],
    'Age': [25, 30, 35, 40, np.nan],
    'Salary': [50000, 54000, 50000, 48000, 47000],
    'Department': ['HR', 'Finance', 'HR', 'Tech', 'Tech']
}

df = pd.DataFrame(data)
print(df)
```

```python
df.info()
```

```python
# Display statistical summary of the DataFrame
print("\nDataFrame Description:")
print(df.describe())
```

```python
# Handle null values by filling with the median (demonstrate another method)
df['Age'].fillna(value=df['Age'].median(), inplace=True)

print("\nDataFrame after filling null values with median:")
print(df)   
```

```python
# Drop rows with any null values if any other columns had nulls (example purpose)
df.dropna(inplace=True)

print("\nDataFrame after dropping any rows with null values:")
print(df)
```

```python
# Create additional department data
department_data = {
    'Department': ['HR', 'Finance', 'Tech'],
    'Budget': [100000, 150000, 200000]
}

df_departments = pd.DataFrame(department_data)

```

```python
# Use apply() to create a new column - 'Seniority' based on Age
df['Seniority'] = df['Age'].apply(lambda x: 'Senior' if x > 30 else 'Junior')

print("\nDataFrame after applying lambda function on Age:")
print(df)
```

```python
# Use a list comprehension to update 'Salary' based on 'Seniority'
df['Salary'] = [x + 5000 if y == 'Senior' else x for x, y in zip(df['Salary'], df['Seniority'])]

print("\nDataFrame after updating Salary using list comprehension:")
print(df)
```

```python
# Merge the initial DataFrame with the department data
df_merged = pd.merge(df, df_departments, on='Department')
print("DataFrame after Merge:")
print(df_merged)
```

```python
# Read data from a JSON format
json_data = '''
[
    {"Name": "Frank", "Age": 28, "Salary": 62000, "Department": "HR"},
    {"Name": "Gina", "Age": 24, "Salary": 58000, "Department": "Tech"}
]
'''
df_from_json = pd.read_json(json_data)
print("\nDataFrame from JSON Data:")
print(df_from_json)
```

```python
# Concatenate JSON data with the existing DataFrame
df_combined = pd.concat([df, df_from_json], ignore_index=True)
print("\nDataFrame after Concatenating JSON data:")
print(df_combined)
```

```python
# Group by 'Department' and summarize the data
grouped_df = df_combined.groupby('Department').agg({
    'Salary': ['mean', 'min', 'max'],
    'Age': ['mean', 'min', 'max']
})
print("\nGrouped DataFrame by 'Department':")
print(grouped_df)
```
